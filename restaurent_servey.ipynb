# **Data Preprocessing**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for column in df.columns:
  if df[column].dtype == np.number:
    continue
  df[column] = LabelEncoder().fit_transform(df[column])
```

```python
df
```

```python
x = df.drop('Covid19 Status' ,axis=1)
```

```python
y = df['Covid19 Status']
```

# Shuffle

```python
from sklearn.utils import shuffle
```

```python
x,y=shuffle(x,y)
```

# **Scaling the data set**

```python
from sklearn.preprocessing import StandardScaler
```

```python
feature_x = StandardScaler().fit_transform(x)
```

```python
feature_x
```

# **Split data set**

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.25 , random_state = 42)
```

```python
xtest
```

# **Build Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
pred
```

# **BaggingRegressor**

```python
from sklearn.ensemble import BaggingRegressor
```

```python
br = BaggingRegressor(n_estimators=1)
br.fit(xtrain, ytrain)
```

```python
pred = br.predict(xtest)
pred
```

# **Performence matrics**

```python
from sklearn.metrics import accuracy_score
```

```python
Accuracy=accuracy_score(ytest,pred)
Accuracy
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
```

```python

cm=confusion_matrix(ytest,pred)
display = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[False,True])
display.plot()
plt.show()
```

```python
from sklearn.metrics import classification_report
import sklearn.metrics as metrics
```

```python
target_names = ['class 1', 'class 2']
print(classification_report(ytest,pred,target_names=target_names))
```

```python
Specificity = metrics.recall_score(ytest, pred, pos_label=0) #True Negative / (True Negative + False Positive), Since it is just the opposite of Recall, we use the recall_score function, taking the opposite position label:
Specificity
```

```python
Sensitivity = metrics.recall_score(ytest, pred) #True Positive / (True Positive + False Negative), Sensitivity is good at understanding how well the model predicts something is positive:
Sensitivity
```

```python
Efficiency = ( Sensitivity + Specificity + Accuracy ) / 3
Efficiency
```

```python
training_accuracy = []
test_accuracy = []

est = range(1, 51)
for n in est:

  br = BaggingRegressor(n_estimators=n)
  br.fit(xtrain, ytrain)

  training_accuracy.append(br.score(xtrain, ytrain))
  test_accuracy.append(br.score(xtest, ytest))

plt.plot(est, training_accuracy, label="training accuracy")
plt.plot(est, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")  
plt.xlabel("n_estimators")
plt.legend()
```
